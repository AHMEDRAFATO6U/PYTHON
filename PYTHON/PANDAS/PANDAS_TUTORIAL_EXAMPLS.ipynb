```python
#7.	Merging and Joining
#  Merge DataFrames based on a common column.
df1 = pd.DataFrame({'EmployeeID': [1, 2, 3, 4],
                    'Name': ['ahmed', 'samy', 'ali', 'alaa'],
                    'Department': ['HR', 'IT', 'Finance', 'IT']})

df2 = pd.DataFrame({'EmployeeID': [2, 3, 4, 5],
                    'Salary': [60000, 45000, 70000, 55000]})


merged_df = pd.merge(df1, df2, on='EmployeeID', how='inner')# (inner): Use intersection of keys from both frames.

print(merged_df)
```

```python
df1 = pd.DataFrame({'EmployeeID': [1, 2, 3, 4],
                    'Name': ['ahmed', 'samy', 'ali', 'alaa'],
                    'Department': ['HR', 'IT', 'Finance', 'IT']})

df2 = pd.DataFrame({'EmployeeID': [2, 3, 4, 5],
                    'Salary': [60000, 45000, 70000, 55000]})


merged_df = pd.merge(df1, df2, on='EmployeeID', how='outer')#outer Use union of keys from both frames, filling in missing values with NaN

print(merged_df)
# look at mee ohhhh 
# look at employeeeid = 5  see it naaaaaaaaan ya fanaan 
```

```python
df1 = pd.DataFrame({'EmployeeID': [1, 2, 3, 4],
                    'Name': ['ahmed', 'samy', 'ali', 'alaa'],
                    'Department': ['HR', 'IT', 'Finance', 'IT']})

df2 = pd.DataFrame({'EmployeeID': [2, 3, 4, 5],
                    'Salary': [60000, 45000, 70000, 55000]})


merged_df = pd.merge(df1, df2, on='EmployeeID', how='left')# (left): Use only keys from the left frame.

print(merged_df)
```

```python
df1 = pd.DataFrame({'EmployeeID': [1, 2, 3, 4],
                    'Name': ['ahmed', 'samy', 'ali', 'alaa'],
                    'Department': ['HR', 'IT', 'Finance', 'IT']})

df2 = pd.DataFrame({'EmployeeID': [2, 3, 4, 5],
                    'Salary': [60000, 45000, 70000, 55000]})


merged_df = pd.merge(df1, df2, on='EmployeeID', how='right')# (right).Use only keys from the left frame

print(merged_df)
```

```python
#•	concat(): Concatenate DataFrames along a particular axis
# Creating two sample DataFrames
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'Key': ['K0', 'K1', 'K2', 'K3']})

df2 = pd.DataFrame({'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3'],
                    'Key': ['K0', 'K1', 'K2', 'K3']})
concatenated_df = pd.concat([df1, df2], axis=0, ignore_index=True)   # 'ignore_index=True' resets the index after concatenation.
print(concatenated_df)
```
